```python
# Path to your CSV file
file_path = "/content/drive/MyDrive/Colab Notebooks/news_dataset.csv"

rows = []
max_columns = 0
bad_lines = []

with open(file_path, 'r', encoding='utf-8', errors='replace') as f:
    for idx, line in enumerate(f):
        # Remove line breaks and strip white space
        cleaned_line = line.strip()

        # Skip empty lines
        if not cleaned_line:
            continue

        # Split using comma (adjust delimiter if needed)
        values = cleaned_line.split('""')

        # Keep track of the widest row (for debugging)
        if len(values) > max_columns:
            max_columns = len(values)

        # Optionally: you can set an expected column count and track malformed lines
        expected_cols = 50  # Change this to expected number of columns
        if len(values) > expected_cols:
            bad_lines.append((idx + 1, len(values), cleaned_line))
            continue

        rows.append(values)


# Create a DataFrame (you may need to provide column names)
df = pd.DataFrame(rows)

# Optional: Print malformed lines to debug
print(f"\nTotal bad lines: {len(bad_lines)}")
for i, (line_num, col_count, content) in enumerate(bad_lines[:5]):
    print(f"Line {line_num} had {col_count} columns: {content[:120]}...")

# Show sample data
print("\nSample data:")
df.head()
```

```python
from datetime import datetime
import pandas as pd

def parse_mixed_dates(date_str):
    if date_str is None:  # Handle None values
        return pd.NaT
    for fmt in ("%B %d, %Y", "%b %d, %Y"):  # Full month, Abbreviated month
        try:
            return datetime.strptime(date_str.strip(), fmt)
        except ValueError:
            continue
    return pd.NaT  # If no format matches

df_date_corrected = df_correct.copy() # Create a copy to avoid modifying the original DataFrame
df_date_corrected['date'] = df_date_corrected['date'].apply(parse_mixed_dates)
display(df_date_corrected.head(10))
```

```python
import numpy as np

# Example: shift left if column '6' is empty string
def shift_row_left_if_empty(row, col=6):
    if row[col] == '':
        # Convert to list, remove the empty cell
        values = row.tolist()
        values.remove('')
        # Pad the end to maintain row length
        values += [np.nan] * (len(row) - len(values))
        return pd.Series(values, index=row.index)
    return row

df_wrong = df_wrong.apply(shift_row_left_if_empty, axis=1)
```
